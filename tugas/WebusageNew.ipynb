## **1. Import Library**

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

## **2. Load & Read Dataset**

```python
nama_file = '/content/drive/MyDrive/PPW/webuage.csv'
df = pd.read_csv(nama_file)
```

```python
display(df)
```

```python
df.info()
```

## **3. Filter Data**

```python
filtered = df[
    (df["request_method"].isin(["GET", "POST"])) &
    (df["request_uri"].isin([
        "/faq.html",
        "/index.html",
        "/somefile.zip",
        "/contribute.txt",
        "/robots.txt"
    ])) &
    (df["status"].isin([200, 301, 404, 206]))
]
```

```python
display(df)
```

## **4. Konversi Waktu ke Format Datetime**

```python
filtered['request_time'] = pd.to_datetime(
    filtered['request_time'],
    errors='coerce',
    utc=True,
    infer_datetime_format=True
)

# Hapus data yang gagal dikonversi (NaT)
df_bersih = filtered.dropna(subset=['request_time'])
```

```python
# Urutkan data berdasarkan IP dan waktu
df_sorted = df_bersih.sort_values(by=['remote_host', 'request_time']).reset_index(drop=True)
```

```python
display(df_sorted)
```

## **5. Identifikasi Sesi User**

### **1. Setiap Sesi**

```python
session_timeout = timedelta(minutes=30)
df_sorted['Session Number'] = 0
df_sorted['Session ID'] = ""

# Hitung sesi per IP berdasarkan selisih waktu antar akses
for ip, group in df_sorted.groupby('remote_host'):
    group = group.sort_values('request_time').copy()
    waktu_akses = group['request_time'].diff().fillna(pd.Timedelta(seconds=0))

    # Jika selisih waktu antar akses > 30 menit → sesi baru
    jumlah_sesi = (waktu_akses > session_timeout).cumsum() + 1

    # Simpan hasil sesi ke dataframe utama
    df_sorted.loc[group.index, 'Session Number'] = jumlah_sesi
    df_sorted.loc[group.index, 'Session ID'] = ip + "_S" + jumlah_sesi.astype(str)

# jumlah sesi = integer
df_sorted['Session Number'] = df_sorted['Session Number'].astype(int)
```

```python
display(df_sorted)
```

### **2. Total Sesi Setiap IP**

```python
total_sesi = (
    df_sorted.groupby('remote_host')['Session ID']
    .nunique()
    .reset_index()
    .rename(columns={'Session ID': 'Total Sessions'})
)

df_baru = pd.merge(df_sorted, total_sesi, on='remote_host', how='left')
```

```python
# Urutkan berdasarkan jumlah sesi terbanyak
df_baru = df_baru.sort_values(
    by=['Total Sessions', 'remote_host', 'request_time'],
    ascending=[False, True, True]
).reset_index(drop=True)
```

```python
display(df_baru)
```

## **6. Kunjungan  Web**

```python
kunjungan_dengan_waktu = (
    df_baru
    .groupby(['remote_host', 'request_time', 'request_uri'])
    .size()
    .unstack(fill_value=0)
    .reset_index()   # supaya request_time muncul sebagai kolom
)
```

```python
# Durasi antar halaman dalam menit
df_baru['Next Time'] = df_baru.groupby(['remote_host', 'Session Number'])['request_time'].shift(-1)
df_baru['Next Request URI'] = df_baru.groupby(['remote_host', 'Session Number'])['request_uri'].shift(-1)

df_baru['Duration (minutes)'] = (
    df_baru['Next Time'] - df_baru['request_time']
).dt.total_seconds() / 60

# Default durasi 30 menit jika tidak ada halaman berikutnya
df_baru['Duration (minutes)'] = df_baru['Duration (minutes)'].fillna(30)
```

```python
durasi_kunjungan = df_baru.pivot_table(
    index='remote_host',
    columns='request_uri',
    values='Duration (minutes)',
    aggfunc='sum',
    fill_value=0
)
```

## **7. Datalog Filter + Durasi**

```python
# data
kolom_df = []
for page in kunjungan.columns:
    kolom_df.append(page)
    kolom_df.append(f"Lama waktu akses di {page} (menit)")

data = pd.DataFrame(index=kunjungan.index)

for page in kunjungan.columns:
    data[page] = kunjungan[page]
    data[f"Lama waktu akses di {page} (menit)"] = durasi_kunjungan[page]

data = data.reset_index()
```

```python
# Tambahkan Total Sessions dari data utama
total_sesi_ip = df_baru[['remote_host', 'Total Sessions']].drop_duplicates()
data = data.merge(total_sesi_ip, on='remote_host', how='left')
```

```python
# Urutkan berdasar Total Sessions
data = data.sort_values(by='Total Sessions', ascending=False).reset_index(drop=True)
```

```python
display(data)
```

```python
# Datalog web setelah pengelompokkan user

# Buat mapping IP unik ke User ID
ip_user = {ip: f"User {i+1}" for i, ip in enumerate(data['remote_host'].unique())}

# Tambahkan kolom 'User' berdasarkan mapping
data['User'] = data['remote_host'].map(ip_user)

# Urutkan supaya kolom 'User' muncul di paling kiri
kolom = ['User'] + [col for col in data.columns if col != 'User']
df_final = data[kolom]

display(df_final)
```

```python
# simpan ke csv
df_final.to_csv('PPW_Tugas7_LogFile(WebUsageMining).csv', index=False)
```

```python
# Mulai dari df_final, ambil remote_host & request_uri
summary_data = df_sorted[['remote_host', 'request_time','request_uri']].copy()

# Buat kolom biner
summary_data['faq.html']       = summary_data['request_uri'].apply(lambda x: 1 if 'faq.html' in x else 0)
summary_data['index.html']     = summary_data['request_uri'].apply(lambda x: 1 if 'index.html' in x else 0)
summary_data['somefile.zip']   = summary_data['request_uri'].apply(lambda x: 1 if 'somefile.zip' in x else 0)
summary_data['contribute.txt'] = summary_data['request_uri'].apply(lambda x: 1 if 'contribute.txt' in x else 0)
summary_data['robots.txt']     = summary_data['request_uri'].apply(lambda x: 1 if 'robots.txt' in x else 0)

# Hapus kolom request_uri karena sudah jadi biner
summary_data_clean = summary_data.drop(columns=['request_uri'])

# Tampilkan hasil
display(summary_data_clean)
```

```python
# Hitung modus berdasarkan halaman yang dikunjungi
modus_per_user = (
    summary_data_clean.groupby(['remote_host'])[['faq.html', 'index.html', 'somefile.zip', 'contribute.txt', 'robots.txt']]
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else 0)
    .reset_index()
)

display(modus_per_user)
```
